#### *What is programmable Flow Injection?* ####

Programmable Flow Injection analysis builds upon previous generations of flow analysis techniques, whereby a target analyte is measured by transporting a sample and reagent solution through an instrument manifold by the use of peristaltic pumps, and quantified via a detector (Cerda et al., 2014). While traditional flow analysis technologies remain extremely useful in analytical chemistry, several limitations exist in these methodologies which are addressed through pFI. One of the key advantages of pFI relative to conventional Flow Injection Analysis (FIA) is the incorporation of a “confluence point”, connecting sample/reagent ports to two bidirectional pumps, allowing for highly efficient mixing using predetermined flow rates and volumes (Hatta et al., 2019). In addition, pFI instrumentation is miniaturized and does not require continuous flow, meaning that analysis can be performed using much smaller sample/reagent volumes, reducing the amount of hazardous waste produced.

The main components of the pFI instrument that I am using includes:
* Two bi-directional peristaltic pumps 
* Thermostatically regulated holding coils
* 6-port Lab-on-Valve equipped with a confluence point
* Tungsten light source attached to a 10cm flow cell and external spectrophotometer via optical fibers

#### *How can pFI be used to measure phosphate in seawater?* ####

Phosphate determination can be done using methods originally described by Murphy & Riley (1962) in which a two step assay is used where acidified molybdate reacts with the phosphate in a sample to form phosphomolybdic acid. The phosphomolybdic acid complex is then reduced to phosphomolybdenum blue using ascorbic acid. The phosphomolybdenum blue is monitored via absorbance spectrophotometry, which is proportional to the amount of phosphate in a sample, based on the Beer-Lambert Law. Phosphate is typically monitored at a wavelength of 880nm. 

This method is applied to pFI by using the mixing sequence shown and described below:
1. Ascorbic Acid is aspirated into Holding Coil 1. 
2. Ascorbic Acid is dispensed while the Molybdate reagent is aspirated into Holding Coil 2 at a 1:1 ratio.
3. The sample is aspirated while the mixed Ascorbic Acid + Molybdate reagent is dispensed into Holding Coil 1.
4. The mixed sample + reagent assay is dispensed to the flow cell for detection after a 30 second wait period.

#### *pFI Data Storage and File Structure* ####

When a sample is run using pFI, the pFI software (FloZF) stores two tab-delimited files into a project's data folder:
1. The **absorbance spectrum** of the sample over the range of wavelengths used by the spetrophotometer. The file header (the first 3 rows) include information about the unique index, or identifier for each sample that is run within a project, as well as teh date and time that the sample was run. The remaining rows include the absorbance values at each wavelength. A typical concentrated phosphate standard has the following absorbance spectrum, with a distinct peak at 880nm.

```python
pfi.plot_spectra(120) #plotting the absorbance spectrum for any sample run with pFI, based on the index/unique identifier of the sample. 
```

2. A **time series** of the absorbance values over the period of sample analysis at wavelengths specified using FloZF. The first 31 lines in the file include recorded information about the absorbance values at specified times in a sequence using various reference wavelengths. The remainder of the file includes the time series of absorbance values using the specified reference wavelengths. In the example below, the absorbance peak is monitored at around 90 seconds, once the mixed sample + reagent solution has been properly mixed and sent to the flow cell for detection.

```python
pfi.plot_timeseries(120,'A880-A510')
```

#### *Analyzing pFI Calibration Data* ####

Running calibrations is critical for any method development, and is ultimately used for calculating concentrations of an analyte in a seawater sample. A linear model seems to fit phosphate calibration well using pFI; however, the residuals using the linear least-squares regression indicates a systematic error where low phosphate concentrations tend to be lower than the model, while higher phosphate concentrations tend to lie above my least-squares regression. A 2nd-order polynomial removes the systematic bias, and is therefor more suitable for calculating phosphate concentrations in seawater samples.

```python
indeces = np.arange(1221,1233) #list of sample indeces used in calibration

pfi.calib_plots(indeces)
```

Some important figures of merit have been included in a function called "calib_stats". This function simply uses the sample indeces as an input, with a defult alpha value set as 0.05. 

```python
pfi.calib_stats(indeces) #calibration statistics using the same calibration as in the calibration plots example above
```

It is useful to identify it outliers exist when running numerous samples at time, whether is be for running a calibration or running real seawater samples as a profile or time-series. The Grubb's test calculates a test statistic "G", based on the suspect value, the mean, and the standard devation of the dataset. Using a critical t-value and the degrees of freedom of the dataset, the hypothesis that there is an outlier can then either by accepted or rejected.  

It should be noted that a Grubb's test is most robust in a dataset containing at least 7 datapoints. For the purpose of this application, most samples are run in triplicates using pFI. Another assumption of this function is that the data is normally distributed; to test this, one could measure a phosphate standard containing a known concentration numerous times to idenfiy the frequency distribution of the absorbance values within the standard. More information on using Grubb's outlier test in analytical chemistry can be found in the book: Statistics and Chemometrics for Analytical Chemistry, Miller et al., 2018. 

```python
#Using the Grubb's outlier test

indeces = np.arange(1221,1233)

###extracting absorbance values from specified indices used in calibration

absorbances = dict()
A880_A510 = dict()
y = []

for number in indeces:
    absorbances[number] = pd.read_csv('master_data/sample_'+str(number)+'.pfl', delimiter='\t', nrows=30)
    absorbances[number] = absorbances[number].replace(to_replace=absorbances[number]['sample name'][23], value='A880-A510', inplace=False, limit=None, regex=False, method='pad')
    absorbances[number] = absorbances[number].replace(to_replace=absorbances[number]['sample name'][16], value='A880-A775', inplace=False, limit=None, regex=False, method='pad')
    absorbances[number] = absorbances[number].replace(to_replace=absorbances[number]['sample name'][9], value='A880-A975', inplace=False, limit=None, regex=False, method='pad')

    Abs = absorbances[number]['sample name'] == "A880-A510" #Extracting absorbance values that only correspond to the 510nm reference wavelength.
    A880_A510[number] = float(absorbances[number][Abs]['sample']) #Completed dictionary which contains absorbance values at 880-510nm for any index.
            
    test = A880_A510[number]
    y.append(test) #list of absorbance values for all specified indeces in input

###slicing list of absorbance values into unique lists for each set of triplicates

blanks = y[:3] #blank standards
rep1 = y[3:6] #0.5uM standards
rep2 = y[6:9] #1.5uM standards
rep3 = y[9:12] #3.0uM standards

reps = [blanks,rep1,rep2,rep3] #list of each replicate list

#idenfiying if any of the replicates run in calibration contain outliers
for i in reps:
    pfi.outlier_test(i) 
```

```python
#Changing one of the blank absorbance values to 0.03 results in the outlier test passing, as shown in the output of the function
blanks[2] = 0.03 #modifying one of the blank values
pfi.outlier_test(blanks)
```

#### *Applying Functions to PO4 Method Optimization: How Does Increasing the Sample:Reagent Ratio Impact the Analytical Sensitivity of this Method?* ####

```python
#Extract calibration slopes using linear regression based on indeces used for calibrations run using each sample dilution ratio
three = pfi.slope_from_index(np.arange(1118,1130)) #3:1 sample:reagent reatio
four = pfi.slope_from_index(np.arange(1097,1109)) #4:1 sample:reagent reatio
five = pfi.slope_from_index(np.arange(1076,1088)) #5:1 sample:reagent reatio
six = pfi.slope_from_index(np.arange(1054,1066)) #6:1 sample:reagent reatio
seven = pfi.slope_from_index(np.arange(1139,1151)) #7:1 sample:reagent reatio
eight = pfi.slope_from_index(np.arange(964,976)) #8:1 sample:reagent reatio

#list of slopes for each dilution factor
y = np.array([three, four, five, six, seven, eight])
```

```python
#Plotting results

#Barplot
x = ['3:1','4:1','5:1','6:1','7:1','8:1']
plt.figure(figsize=[18,5])
plt.subplot(1,2,1)
plt.bar(x,y,width=0.5,color='coral')
plt.ylim(0.08,0.13)
plt.xlabel('Sample:Reagent Ratio')
plt.ylabel('Calibration Slope [uM PO4/mAU]')
plt.title('Sample:Reagent Ratio vs Analytical Sensitivity (A880-A510)')
plt.grid(axis='y')


#Scatterplot with linear regression
x = np.arange(3,9)
plt.subplot(1,2,2)
plt.plot(x,y,'.')

lin_results = stats.linregress(x,y)
a2 = lin_results[0] 
intercept = lin_results[1]
y_hat = (a2*x) + intercept

plt.plot(x,y_hat,'-')

plt.xlabel('Reagent Dilution Factor')
plt.ylabel('Calibration Slope (mAU/uM PO4)')
plt.title('Sample:Reagent Ratio vs Analytical Sensitivity (A880-A510)')
```

Increasing the volume of sample that is mixed with the reagents tends to increase the sensitivity of the method, indicated by an increase in the calibration slope. This is due to the fact that as the sample:reagent ratio is increased, there is a larger volume of sample that can react with the acidifed molybate and ascorbic acid, forming a greater quantify of phosphomolybedum blue, thus increasing the sensitivity of the method. This relation tends to be linear, despite some variability in the calibration slopes that could be associated with inconsistencies in mixing, slightly different chemical makeup of reagents, etc.

#### *Quantifying PO4 in Monterey Bay* ####

As shown above, using a polynomial model reduces any systematic bias. Optimized reagent composition and instrumentation conditions were used to validate this pFI method using Certified Reference Materials (results not included). This phosphate benchtop method was then used to measure phosphate concentrations in Monterey Bay from the surface to 900dbar. 

```python
df = pd.read_csv('MBprofile.csv') #importing file with list of indeces corresponding to the depth at which the sample was collected

calib_indeces = np.arange(1386,1398) #list of indeces used in calibration 
sample_indeces = df['index'] #list of indeces used for measuring Monterey Bay profile

df['PO4'] = pfi.solve_for_conc(calib_indeces,sample_indeces) #Adding a column in dataframe with calculated concentrations using polynomial calibration model

df2 = df.groupby(np.arange(len(df))//3).mean() #Calculating the mean concentration of each set of triplicate samples to ensure only one datapoint corresponds to each depth

#Plotting Monterey Bay profile 
fig, ax = plt.subplots(figsize=[8,8])
plt.plot(df2['PO4'],df2['dbar'],'.',ms=8)
plt.gca().invert_yaxis()
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')
plt.grid()
plt.xlabel('PO4 [uM]')
plt.ylabel('Pressure [dbar]')
plt.title('Montere Bay PO4 Profile',fontsize=15)
```

The above Monterey Bay PO4 profile shows an increase in phosphate with depth; this trend is typical in many parts of the ocean. PO4 is assimilated by phytoplankton near the surface, drawing down the concentration in the top ~100m, and increases with depth as particulate organic matter containing phosphate sinks to depth. It should be noted that phosphate profiles usually have a smoother curve; the high variability in this profile is likely associated with issues in the sample preservation method.
